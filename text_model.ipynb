```python
import pandas as pd
df = pd.read_csv('C:/Users/tiger/Downloads/musinsa_recommendation_system/데이터_전처리_완료.csv'); df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

text_columns = ['태그', '카테고리', '리뷰']

for col in text_columns:
    print(col,"유사도")
    tfidf = TfidfVectorizer(min_df=1, ngram_range=(1, 3))
    matrix = tfidf.fit_transform(df[col])
    cos_sim = cosine_similarity(matrix, matrix)
    print(cos_sim[5:16, 5:16]) # 행렬의 특정 행/열만 출력
    print("\n")
```

```python
def tfidf_cos_sim(col):
    tfidf = TfidfVectorizer(min_df=1, ngram_range=(1, 3))
    matrix = tfidf.fit_transform(df[col])
    cos_sim = cosine_similarity(matrix, matrix)
    print(cos_sim)
    return cos_sim

tag = tfidf_cos_sim('태그')
category = tfidf_cos_sim('카테고리')
review = tfidf_cos_sim('리뷰')
brand = tfidf_cos_sim('브랜드')
```

```python
total_sim = ( review * 0.1 + category * 1 + tag * 1 )
            #+ np.repeat(np.repeat([df['평점'].values], len(df['평점']) , axis=0) * 0.001
            #+ np.repeat(np.repeat([df['조회수'].values], len(df['조회수']) , axis=0) * 0.001
            #+ np.repeat(np.repeat([df['구매수'].values], len(df['구매수']) , axis=0) * 0.001 )

total_sim_sorted_ind = total_sim.argsort()[:, ::-1] 

def find_sim_item(df, sorted_ind, item_name, top_n):
    # 입력된 item_name을 포함하는 이름을 가진 아이템을 데이터 프레임에서 찾기
    item_name_df = df[df['상품명'].str.contains(item_name, case=False, na=False)]
    
    # 아이템의 첫 번째 인덱스 값을 item_index에 저장
    item_index = item_name_df.index.values[0] # 첫 번째 인덱스만 사용

    # sorted_ind에서 해당 아이템과 가장 유사한 상품들의 인덱스 추출
    similar_indexes = sorted_ind[item_index, :top_n]
    return df.iloc[similar_indexes]

find_sim_item(df, total_sim_sorted_ind, '컴프레션 긴팔 티셔츠', 10) 
```
