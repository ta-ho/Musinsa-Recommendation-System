```python
df = df[df['카테고리'] != '기타 상의']
```

```python
df['카테고리'] = df['카테고리'].str.replace("/", " ") 
df['카테고리']
```

```python
# 인기 성별
# 남성 => 남
# 여성 => 여
# 없음 => 제거
```

```python
df['인기 성별'].value_counts()
```

```python
df = df[df['인기 성별'] != '없음']

df['인기 성별'] = df['인기 성별'].replace('남성', '남')
df['인기 성별'] = df['인기 성별'].replace('여성', '여')
```

```python
# 성별
# 남\n여 => 공용
# / => 인기 성별과 같도록
# 라이프 => 삭제
```

```python
df['성별'].unique()
```

```python
df['성별'].value_counts()
```

```python
# "라이프" 삭제
df = df[df['성별'] != '라이프']

# "남\n여"를 "공용"으로 변경
df['성별'] = df['성별'].replace('남\n여', '공용')

# "/"를 해당 행의 "인기 성별"로 변경
slash = df['성별'] == '/'
df.loc[slash, '성별'] = df.loc[slash, '인기 성별']
```

```python
df['성별'].value_counts()
```

```python
# 리뷰수
# 글자 삭제 후 숫자만 추출, int형으로
```

```python
df['리뷰수'].unique()
```

```python
import re

def extract_number(s):
    return int(re.sub(r'\D', '', s))

df['리뷰수'] = df['리뷰수'].apply(extract_number)
```

```python
df['리뷰수'].unique()
```

```python
df['태그'] = df['태그'].str.replace("#", " ") 
```

```python
df['태그'].unique()
```

```python
# 좋아요 수
# 쉼표 제거 후 int형으로 
```

```python
df['좋아요 수'].unique()
```

```python
df['좋아요 수'] = df['좋아요 수'].str.replace(',', '').astype(int)
```

```python
df['좋아요 수'].value_counts()
```

```python
# 조회수
# ~만 회 이상 => 텍스트 제거 후 x10000
# ~천 회 이상 => 텍스트 제거 후 x1000
```

```python
df['조회수'].unique()
```

```python
def convert_view_count(value):
    if '천 회 이상' in value:
        return float(value.replace('천 회 이상', '').strip()) * 1000
    elif '만 회 이상' in value:
        return float(value.replace('만 회 이상', '').strip()) * 10000
    elif '회 이상' in value:  # 기타 경우 (예: '500회 이상')
        return float(value.replace('회 이상', '').strip())
    else:
        return value

df['조회수'] = df['조회수'].apply(convert_view_count).astype(int)
df['조회수'].unique()
```

```python
# 구매수
# 조회수와 동일 
# "~개 미만"만 추가로 처리
```

```python
df['구매수'].unique()
```

```python
# "없음" 삭제
df = df[df['구매수'] != '없음']
```

```python
def convert_purchase_count(value):
    if '천 개 이상' in value:
        return float(value.replace('천 개 이상', '').strip()) * 1000
    elif '만 개 이상' in value:
        return float(value.replace('만 개 이상', '').strip()) * 10000
    elif '개 이상' in value: 
        return float(value.replace('개 이상', '').strip())
    elif '개 미만' in value:
        return float(value.replace('개 미만','').strip())
    else:
        return value

df['구매수'] = df['구매수'].apply(convert_purchase_count).astype(int)
df['구매수'].unique()
```
